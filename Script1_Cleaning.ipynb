```python
# dictionary mapping neighbourhood with attraction
dmap_locations = {'Brooklyn' : 'Brooklyn Bridge', 'Manhattan' : 'Central Park', 'Queens' : 'Citi Field', 'Staten Island': 'St. Georges Ferry Terminal', 'Bronx': 'Bronx Zoo'}
```

```python
# dictionary mapping neighbourhood with attraction coordinates
dmap_coordinates = {'Brooklyn': brooklyn_bridge, 'Manhattan': central_park, 'Queens': citi_field, 'Staten Island': staten_island_ferry, 'Bronx': bronx_zoo}
```

```python
df_mapped['attraction_location'] = df_mapped['neighbourhood_group'].map(dmap_locations)
```

```python
df_mapped['attraction_coordinates'] = df_mapped['neighbourhood_group'].map(dmap_coordinates)
```

```python
df_mapped.head()
```

```python
df_distance = df_mapped.copy()
```

```python
df_distance.head()
```

```python
df_distance.dtypes
```

```python
df_distance['distance_to_attraction'] = df_distance.apply(lambda x: haversine(x.listing_coordinates, x.attraction_coordinates, unit = Unit.MILES), axis = 1)
df_distance['distance_to_attraction'] = df_distance['distance_to_attraction'].round(2)
```
